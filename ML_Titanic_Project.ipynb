```python
import numpy as np
import pandas as pd
train_data = pd.read_csv("train.csv")
train_data_noNaN = train_data.dropna(axis=1)
train_data.head()
```

```python
test_data = pd.read_csv("test.csv")
test_data_noNaN = test_data.dropna(axis=1)
test_data.head()
```

```python
#If training with all features doesn't containing NaN
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
features = test_data_noNaN.columns[1:-1]
X = pd.get_dummies(train_data[features])
y = train_data["Survived"]
X_train, X_val, y_train, y_val = train_test_split(X, y)
model = RandomForestClassifier(n_estimators=1000, max_depth=20, random_state=1)
model.fit(X_train, y_train)
predictions = model.predict(X_val)
score = accuracy_score(predictions,y_val)
print('All features without NaN evaluation:',score)
```

```python
import numpy as np
train_data = pd.read_csv("train.csv")
features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_data[features])
y = train_data["Survived"]
X_train, X_val, y_train, y_val = train_test_split(X, y)

from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
candidate_max_leaf_nodes = np.arange(2, 1001, 1, dtype=int)
def get_mae(max_leaf_nodes, X_train,X_val, y_train,y_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(X_train, y_train)
    preds_val = model.predict(X_val)
    mae = mean_absolute_error(y_val, preds_val)
    return(mae)
scores = {i:get_mae(i,X_train,X_val,y_train,y_val) for i in candidate_max_leaf_nodes}
best_tree_size = min(scores, key=scores.get)
model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=0)
model.fit(X_train, y_train)
preds_val = model.predict(X_val)
mae = mean_absolute_error(y_val, preds_val)
print('With DecisionTreeRegressor Model, score = :', 1-mae,'with tree nodes',best_tree_size)
```
